# 🩺 Diabetes Prediction with Feature Engineering

## Project Overview
In this project, we aim to predict whether a person has diabetes based on medical features.

We perform:
- Exploratory Data Analysis (EDA)
- Data preprocessing
- Feature engineering
- Model building

The goal is to improve model performance through feature engineering.

## 📊 Dataset Information

- 768 observations
- 8 numerical features
- Target variable: Outcome (0 = No Diabetes, 1 = Diabetes)

This dataset contains medical attributes of Pima Indian women.

## 🔍 Task 1: Exploratory Data Analysis (EDA)

We analyzed:
- General structure of the dataset
- Variable types
- Distribution of target variable
- Relationships between variables
- Outliers and missing values

## ⚙️ Task 2: Feature Engineering

We improved the dataset by:
- Handling missing values
- Treating outliers
- Creating new meaningful features
- Encoding categorical variables
- Scaling numerical features

## 🤖 Model Results

- Base model and final model were compared
- Feature engineering improved performance
- Random Forest was used for classification

## 🚀 Final Conclusion

- Data preprocessing significantly improved model performance
- Feature engineering revealed hidden patterns in the data
- Glucose, BMI, and Age were the most important features
- The final model achieved strong ROC-AUC performance

This project demonstrates an end-to-end machine learning pipeline.
